```python
import matplotlib.pyplot as plt

# Data fra studien, kilde: https://www.nature.com/articles/nrcardio.2015.35
categories = ['SCD', 'Fatal CHD', 'Fatal CVD', 'All-cause Mortality']
sauna_frequencies = ['Once per week', '2-3 times per week', '4-7 times per week']
HR_values = {
    'Once per week': [1.0, 1.0, 1.0, 1.0],  # HR values for each health outcome
    '2-3 times per week': [0.78, 0.77, 0.73, 0.76],
    '4-7 times per week': [0.37, 0.52, 0.50, 0.60]
}

# Lager en line Chart for hver helse utbytte. 
for i, category in enumerate(categories):
    plt.figure()  # Lager en separat figure for hver helse resultat.
    for frequency in sauna_frequencies:
        # Passe på at HR(heartrate) verdiene matcher sauna_frequencies. 
        plt.plot(sauna_frequencies, HR_values[frequency][:len(sauna_frequencies)], label=frequency)

    plt.xlabel('Sauna Frequency')
    plt.ylabel('Hazard Ratio (HR)')
    plt.title(f'Hazard Ratios for {category}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

plt.show()
```
