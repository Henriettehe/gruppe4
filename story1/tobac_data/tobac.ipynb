```python
import os
import pandas as pd

# Get the absolute path to the CSV file
current_directory = os.getcwd()
relative_path = './HEALTH_LVNG_18112023145804261.csv'
absolute_path = os.path.join(current_directory, relative_path)

# Read the CSV file
df = pd.read_csv(absolute_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Get a list of unique countries in the DataFrame
countries = df['Country'].unique()

# Convert 'Year' column to integer type
df['Year'] = pd.to_numeric(df['Year'], errors='coerce').astype('Int64')

# Limit the DataFrame to include only the year 2021
df_2021 = df[df['Year'] == 2021]

# Plotting
plt.figure(figsize=(12, 8))

for country in countries:
    country_data = df_2021[(df_2021['Country'] == country) & (~df_2021['Value'].isna())]
    plt.bar(country_data['Country'], country_data['Value'], label=country)

plt.title('Tobacco Consumption in 2021 - Selected Countries')
plt.xlabel('Country')
plt.ylabel('Tobacco Consumption (Grams per capita)')
plt.grid(axis='y')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Place legend outside the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Filter data for the year 2021
data_2021 = df[df['Year'] == 2021]

# Update Sweden's value to 54.7
data_2021.loc[data_2021['Country'] == 'Sweden', 'Value'] = 54.7

# Define colors for each country
colors = {
    'Denmark': 'blue',
    'Finland': 'green',
    'Iceland': 'orange',
    'Norway': 'red',
    'Sweden': 'purple',
}

# Plotting
plt.figure(figsize=(12, 8))
bars = plt.bar(data_2021['Country'], data_2021['Value'], color=[colors[c] for c in data_2021['Country']])
plt.title('Tobacco Consumption in 2021 (Grams per capita)')
plt.xlabel('Country')
plt.ylabel('Tobacco Consumption (Grams per capita)')

# Add legend
legend_labels = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in colors]
plt.legend(legend_labels, colors.keys(), loc='upper left')

# Add source text
plt.text(0.5, -0.1, 'Source: OECD Statistics (https://stats.oecd.org/index.aspx?queryid=30127#)', ha='center', va='center', transform=plt.gca().transAxes, color='gray', fontsize=8)

plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Filter data for the year 2021 and the "Grams per capita" measure
data_2021_grams = df[(df['Year'] == 2021) & (df['Measure'] == 'Grams per capita (15+)')]

# Update Sweden's value to 54.7
data_2021_grams.loc[data_2021_grams['Country'] == 'Sweden', 'Value'] = 54.7

# Filter data for the total grams for Sweden
data_2021_grams_total = df[(df['Year'] == 2021) & (df['Measure'] == 'Total') & (df['Country'] == 'Sweden')]
data_2021_grams_total['Country'] = 'Sweden (Total)'

# Filter data for Iceland and Finland
data_2021_grams_isl_fin = df[(df['Year'] == 2021) & (df['Measure'] == 'Grams per capita (15+)') & (df['Country'].isin(['Iceland', 'Finland']))]

# Concatenate the data frames
data_2021_combined = pd.concat([data_2021_grams, data_2021_grams_total, data_2021_grams_isl_fin])

# Define colors for each country
colors = {
    'Denmark': 'blue',
    'Finland': 'green',
    'Iceland': 'orange',
    'Norway': 'red',
    'Sweden': 'purple',
    'Sweden (Total)': 'darkcyan',
}

# Plotting
plt.figure(figsize=(12, 8))
bars = plt.bar(data_2021_combined['Country'], data_2021_combined['Value'], color=[colors[c] for c in data_2021_combined['Country']])
plt.title('Tobacco Consumption in 2021 (Grams per capita and Total)')
plt.xlabel('Country')
plt.ylabel('Tobacco Consumption (Grams)')

# Add legend
legend_labels = [plt.Rectangle((0, 0), 1, 1, color=colors[label]) for label in colors]
plt.legend(legend_labels, colors.keys(), loc='upper left')

# Add source text
plt.text(0.5, -0.1, 'Source: OECD Statistics (https://stats.oecd.org/index.aspx?queryid=30127#)', ha='center', va='center', transform=plt.gca().transAxes, color='gray', fontsize=8)

plt.xticks(rotation=45, ha='right')
plt.show()
```
